```python
import csv
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataUser = pd.read_csv('UserCentreInteret.csv') #Les données de l'utilisateur
dataInfluencer = pd.read_csv('influencersFitness.csv') #Liste des influenceurs qui présente l'intérêt de cet utilisateur 


```

## Critère 1 : Influenceurs qui sont estimés similaires à des influenceurs suivis par l’utilisateur u

```python
#on va chercher les influenceurs qui sont déjà suivis par l'utilisateur u en calculant la distance de similarité entre 
#la liste des followings de l'utilisateur et la liste des influenceurs à recommander 

followingUser = dataUser["Following"] #la liste des followings

distanceFollowersFollowing = [] #array pour stocker les distances entre les followings de u et chaque influenceur de la liste
                                #à recommander
    
vec = CountVectorizer() #Convert a collection of text documents to a matrix of token counts

k = []
k.append(followingUser[0])
f = vec.fit_transform(k).todense()

index=0
listInf = dataInfluencer["username"] #Liste des usernames des influenceurs à recommander

#On parcoure la liste et on calcule à chaque fois la distance de similarité entre username de l'influenceur i et la liste
#des following de l'utilisateur u


while index<len(dataInfluencer):
    c = []
    c.append(listInf[index])
    f1 = vec.transform(c).todense()
    distanceFollowersFollowin.append(cosine_similarity(f,f1))
    index+=1

distanceFollowersFollowin #afficher ces distances

#On peut remarquer que le premier et le 3ème influenceur présente une distance non nulle ce qui prouve
#que ces influenceurs sont déjà suivis par l'utilisateur u
```

```python
a = [] #array pour stocker les indices des influenceurs déjà suivis

i = 0 
while i<len(distanceFollowersFollowin):
    if (distanceFollowersFollowin[i][0][0]!= 0):
        a.append(i)
    i+=1
    
#Maintenant on parcoure la liste des influenceurs à recommander différents des influenceurs déjà suivis 
#et on calcule leurs distances de similarité avec ces derniers
#pour chaque influenceur à recommander on ajoute sa distance de similarité avec les influenceurs déjà suivis à son score
index1=0
while index1<len(a):
    ve = CountVectorizer()
    c = []
    c.append(dataInfluencer["posts__caption"][ a[index1] ])
    f = ve.fit_transform(c).todense()
    index2 = 0
    while index2<len(dataInfluencer):
        if (index2 == a[index1] ):
            dataInfluencer["Score"][index2] = 0
        if (dataInfluencer["Score"][index2] !=0 ):    
            k= []
            k.append(dataInfluencer["posts__caption"][index2])
            f1 = ve.transform(k).todense()
            d = cosine_similarity(f,f1)
            dataInfluencer["Score"][index2] = d   
        index2+=1
    index1+=1
 
```

```python
#On élimine les influenceurs déjà suivis
i=0
while i<len(dataInfluencer):
    if (dataInfluencer['Score'][i]==0):
        dataInfluencer.drop(i,inplace=True)
    i+=1
    
```

```python
#on met à jour les indices de dataset
i=0 
index = []
while i<len(dataInfluencer):
    index.append(i)
    i+=1
    
dataInfluencer.set_index([pd.Index(index)],inplace=True)
```

## Critère 2 : Influenceurs qui présentent les intérêts d’un utilisateur u

```python
#on récupère les # des influenceurs et on les stocke dans l'array corpus
corpus = []

i=0
while i<len(dataInfluencer):
    corpus.append(dataInfluencer["posts__caption"][i])
    i+=1

corpus

```

```python
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(corpus).todense()
print(vectorizer.vocabulary_)
```

```python
user = dataUser["posts__caption"] #on récupère les # de l'utilisateur
features1 = vectorizer.transform(user).todense()
print(features1)
```

```python
print(euclidean_distances(features1[0],features))
```

```python
print(cosine_similarity(features1[0],features))
```

```python
distances = cosine_similarity(features1[0],features)
#Après le calcul de la distance de similarité entre chaque influenceur et l'utilisateur u ,on ajoute cette distance au score
j=0
while j<len(dataInfluencer):
    dataInfluencer["Score"][j] += distances[0][j]
    j+=1
    
```

## Critère 3 : Influenceurs avec le nombre de followers le plus élevé

### Calcul min et max followers__count

```python
c=1
max = dataInfluencer["followers__count"][0]
min = dataInfluencer["followers__count"][0]
while c<len(dataInfluencer):
    if(dataInfluencer["followers__count"][c]>max):
        max = dataInfluencer["followers__count"][c]
    if(dataInfluencer["followers__count"][c]<min):
        min = dataInfluencer["followers__count"][c]    
    c+=1
    
```

### Calcul du taux de réputation de chaque influenceur 

```python
i=0
while (i<len(dataInfluencer)):
    dataInfluencer["Score"][i] += (dataInfluencer["followers__count"][i]-min)/(max-min)
    i+=1
```

## Critère 4 : Influenceurs qui sont suivis par les amis de u

```python
followingUser = dataUser["Following"] #on récupère la liste des amis de u
followingUser[0]
```

```python
#on parcoure la liste des followers de chaque influenceur à recommander et on calcule sa distance de similarité avec la 
#la liste d'amis de u 
distanceFollowers = []
vect = CountVectorizer()
k = []
k.append(followingUser[0])
feat = vect.fit_transform(k).todense()

index=0
followersInf = dataInfluencer["followers"]
while index<len(dataInfluencer):
    c = []
    c.append(followersInf[index])
    fea = vect.transform(c).todense()
    distanceFollowers.append(cosine_similarity(feat,fea))
    index+=1
```

```python
distanceFollowers
```

```python
# pour chaque influenceur on ajoute sa distance au score
c=0
while c<len(dataInfluencer):
    dataInfluencer["Score"][c] += distanceFollowers[c][0][0]
    c+=1
```

# Score final

```python
#pour chaque influenceur On divise son score par 4
c=0
while c<len(dataInfluencer):
    dataInfluencer["Score"][c]=dataInfluencer["Score"][c]/4
    c+=1
```
